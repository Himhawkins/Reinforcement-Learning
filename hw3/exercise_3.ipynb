```python
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
import pandas as pd
import copy
```

### Exercise 2 DP

```python
def x_next(x,u):
    if u-x+1 > 2:
        return 2
    elif u-x+1 <=2 and u-x+1 >=-2:
        return u-x+1
    else:
        return -2
def x(i):
    return i-2

def cost_update(X_space,U_space,Steps=3):
    # initialise the steps matrix and the initial terminal cost 
    X={key:{'_x_Start':key,'u':[],'x':[key],'j':key**2,'j_ctg':[key**2]} for key in X_space}

    # Iterate through Stems
    for k in range(Steps):
        X_old=copy.deepcopy(X) #NESTED OBJECTS WERE NOT BEING COPIED!!
        
        for i in X_space:
            #if there are already enough steps appended, skip update to this x. 
            
            ctg={}
            for j in U_space:
                ctg[j] =  2*abs(i)+abs(j) +X_old[x_next(i,j)]['j']
            # print(f"Example Minimising: For Step {k} (from end) at x={i} , we choose best action by minimising the following g(x,u)+J(x_n+1)")
            # print(ctg)
            j_min=min(ctg, key=ctg.get) 
            
            # print("We choose Action: ",j_min)
            # print(j_min)  
            xn=x_next(i,j_min) 
            X[i]['j']=ctg[j_min]
            X[i]['u']= [j_min]+X_old[xn]['u']
            X[i]['x']= [i]+X_old[xn]['x']    
            X[i]['j_ctg']= [ctg[j_min]]+X_old[xn]['j_ctg']  
    return X

```

#### x_n+1 =     u_n + 1 - x_n if un-xn between -3 , 1;
####            2              if un-xn > 1;
####            -2 otherwise


### initial cost = x3^2 
## cost to go : min(2*modx +mod u + J(xn+1))

we are taking two steps 

```python
x= [-2,-1,0,1,2]
u=[-1,0,1]
Xr=cost_update(x,u)
df = pd.DataFrame.from_dict(Xr, orient='index').reset_index(drop=True)
print("All the controls, for corresponding starting position u, path x, total cost j are as follows: ")
print(df)
```

### Sequence of actions for corresponding start postion can be looked up from the tabular results
* x0 = 0, u =[-1,-1, -1]
* x0 =−2, u= [-1, 1, -1]
* x0 = 2, u= [1, -1, -1]


### Steps for the Algorithm:
1. First the Cost matrix is initialised with terminal Cost (assuming the agent is at X when terminal) x^2. This Cost matrix is 0 step Cost.
2. Then we move one step back, using Bellman's principal of optimality, to X_(n-1) by minimising g(x_n-1,u_n-1)+J(x_n) over all u , where g is abs(x_n-1)+u_n-1
3. Once we have the optimal command and new optimal cost, we update the Cost matrix for all X, and note down U, (note down all Costs to go as well). This Cost matrix is now Cost for a 1 Step Problem.
4. We repeat step 2 and 3 until we get cost matrix for required number of steps =3, and our noted down U list size is also 3. 
5. This Cost is cost of performing 3 steps starting from this X, and applying the the U noted down, we can get the trajectory X as well.
